```python
dados_ibge = pd.read_excel('POP2022_Municipios.xls', header=1, skipfooter=34)
```

```python
dados_ibge
```

```python
dados_ibge.groupby('UF').sum(numeric_only=True)
```

```python
dados_ibge[dados_ibge['POPULAÇÃO'].str.contains('\(', na = False)]
```

```python
dados_ibge = dados_ibge.assign(parenteses = dados_ibge['POPULAÇÃO'].replace('\(\d{1,2}\)', '', regex = True), populacao = lambda x: x.loc[:, 'parenteses'].replace('\.','',regex=True))
```

```python
dados_ibge[dados_ibge['POPULAÇÃO'].str.contains('\(', na = False)]
```

```python
dados_ibge = dados_ibge.astype({'populacao':'int64'})
dados_filtrados =dados_ibge.groupby('UF')[['populacao']].sum().reset_index()
```

```python
dados_estado = dados_ano[dados_ano['Ano'] == 2021].groupby('Estado')[['Emissão']].sum()
dados_estado
```

```python
dados_agrupados = pd.merge(dados_filtrados, dados_estado, left_on='UF', right_on='Estado')
dados_agrupados
```

```python
dados_agrupados = dados_agrupados.assign(per_capita = dados_agrupados['Emissão']/dados_agrupados['populacao']).sort_values('per_capita', ascending=False)
dados_agrupados
```
