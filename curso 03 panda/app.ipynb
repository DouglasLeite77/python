```python
dados = dados[dados['Emissão / Remoção / Bunker'] == 'Emissão']
dados.head()
```

```python
dados = dados.drop(columns='Emissão / Remoção / Bunker')
dados.head()
```

```python
columns_info = dados.loc[:,'Nível 1 - Setor':'Produto'].columns
columns_info
```

```python
anos_dados = dados.loc[:,1970:2021].columns
anos_dados
```

```python
dados.melt(id_vars=columns_info, value_vars=anos_dados, var_name='Ano', value_name='Emissão')
```

```python
dados_ano = dados.melt(id_vars=columns_info, value_vars=anos_dados, var_name='Ano', value_name='Emissão')
dados_ano.head()
```

```python
dados_ano.groupby('Gás').sum(numeric_only='True')
```

```python
dados_gás = dados_ano.groupby('Gás')[['Emissão']].sum().sort_values('Emissão',ascending=False)
dados_gás
```

```python
dados_gás.iloc[0:9]
```

```python
print(f'{float((dados_gás.iloc[0:9].sum()/dados_gás[:].sum()).iloc[0])*100:.2f}')
```

```python
print(f'A emissão de CO2 corresponde a {float((dados_gás.iloc[0:9].sum()/dados_gás.sum()).iloc[0])*100:.2f} % de emissão total de gases estufa no Brasil de 1970 a 2021.')
```
